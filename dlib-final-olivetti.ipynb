```python
def calc(x1,x2,y1,y2):
    v1=[x1,x2]
    v2=[y1,y2]
    mx=LA.norm(v1)
    my=LA.norm(v2)
    #print mx,my
    a=x1/mx
    b=x2/mx
    c=y1/my
    d=y2/my
    dist=(((a-b)**2)+((c-d)**2))
    s=numpy.sqrt(dist)
    return dist
```
