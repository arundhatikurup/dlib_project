```python
import cv2 , os
import dlib
import numpy 
import matplotlib.pyplot as plt
from PIL import Image
from numpy import linalg as LA
import time
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
PREDICTOR_PATH = "/home/arundhati/cv/dlib_project/shape_predictor_68_face_landmarks.dat/data"

predictor = dlib.shape_predictor(PREDICTOR_PATH)
detector = dlib.get_frontal_face_detector()
```

```python
def get_landmarks1(im):
    try:
        rects = detector(im, 1)	
        a=numpy.matrix([[p.x, p.y] for p in predictor(im, rects[0]).parts()])
    except IndexError :
        pass

    else :
        return a
```

```python
def get_landmarks(im):
    try:
        rects = detector(im, 1)	
        a=numpy.array([[p.x, p.y] for p in predictor(im, rects[0]).parts()])
    except IndexError :
        pass

    else :
        return a
```

```python
def scale(x1,x2,y1,y2):
    v1=[x1,x2]
    v2=[y1,y2]
    mx=LA.norm(v1)
    my=LA.norm(v2)
    #print mx,my
    a=x1/mx
    b=x2/mx
    c=y1/my
    d=y2/my
    #print a,b,c,d
    return a,b,c,d

def calculate(landmarks):
    c=[]
    #co=0
    for point in (landmarks):
        pos=(point[0,0],point[0,1])
    #co=co+1
        c.append(pos)
    #print co
    #print c

    combn=[]
    for i in range(0,68,1):
        for j in range(0,68,1):
            #print c[i][0],c[j][0],c[i][1],c[j][1]
            x1,x2,y1,y2=scale(c[i][0],c[j][0],c[i][1],c[j][1])
            #print x1,x2,y1,y2
            #d=(((c[i][0]-c[j][0])**2)+((c[i][1]-c[j][1])**2))
            d=(((x1-x2)**2)+((y1-y2)**2))
            s=numpy.sqrt(d)
            combn.append(s)
    return combn
    #print len(combn)-each is 68*68
    #print combn[10],combn[100],combn[23],combn[37]
```

```python
def calc(x1,x2,y1,y2):
    v1=[x1,x2]
    v2=[y1,y2]
    mx=LA.norm(v1)
    my=LA.norm(v2)
    #print mx,my
    a=x1/mx
    b=x2/mx
    c=y1/my
    d=y2/my
    dist=(((a-b)**2)+((c-d)**2))
    s=numpy.sqrt(dist)
    return dist
```

```python
final=[]
l=[]

image_rootDir="/home/arundhati/cv/dlib_project/orl_faces"
for dirName, subdirList, fileList in os.walk(image_rootDir):
        for fname in fileList:
            if fname.endswith(('.pgm')):
                #feature=[]
                label=os.path.basename(dirName)
                image_path = os.path.join(dirName, fname)
                image_pil = Image.open(image_path).convert('L')
                image = numpy.array(image_pil, 'uint8')
                try:
                    landmarks1=get_landmarks1(image) 
                    feature=calculate(landmarks1)
                except TypeError: 
                    print fname
                    continue
                   
                l.append(label)    
                final.append(feature)

print len(final)
print len(final[0])
print len(l)
print final[0]
```

```python
print l[]
```

```python
from sklearn.preprocessing import Imputer
X = Imputer().fit_transform(final)
```

```python
forest = ExtraTreesClassifier(n_estimators=250,random_state=0)
forest.fit(X,l)
```

```python
importances = forest.feature_importances_
```

```python
indices = numpy.argsort(importances)[::-1]
print len(indices)
```

```python
pq=[]
for i in range(0,200,1):
    a=indices[i]
    p1=a/68
    p2=a%68
    a=numpy.matrix([p1,p2])
    #v=[p1]+[p2]
    pq.append(a)
print pq    
        
```

```python
final1=[]
l1=[]
image_rootDir="/home/arundhati/cv/dlib_project/orl_faces"
for dirName, subdirList, fileList in os.walk(image_rootDir):
        for fname in fileList:
            if fname.endswith(('.pgm')):
                p1=[]
                label=os.path.basename(dirName)
                image_path = os.path.join(dirName, fname)
                image_pil = Image.open(image_path).convert('L')
                image = numpy.array(image_pil, 'uint8')
                try:
                    landmarks=get_landmarks(image)
                    #print landmarks1
                    for point in pq:
                        a=point[0,0]
                        b=point[0,1]
                        x1,y1=landmarks[int(a)]
                        x2,y2=landmarks[int(b)]
                        #l.append(label)
                        dist1=calc(x1,x2,y1,y2) 
                        p1.append(dist1)
                except TypeError: 
                    print fname
                    continue
                l1.append(label)    
                final1.append(p1)

print len(final1)
print len(final1[0])
print len(l1)
print final1[0]

```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=20)
pca.fit(final1)
X_reduced = pca.transform(final1)
print("Reduced dataset shape:", X_reduced.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_reduced,l1, test_size=0.33, random_state=42)
knn = KNeighborsClassifier(n_neighbors=3, weights='distance')
knn.fit(X_train,y_train)
```
